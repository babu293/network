```python
import cv2
import numpy as np
```

```python
img = cv2.imread("D:/babu/dog.jpg")
```

```python
# Random horizontal flip
def horizontal_flip(img):
    return cv2.flip(img, 1)

# Random rotation
def rotation(img, angle_range=15):
    rows, cols = img.shape[:2]
    angle = np.random.uniform(-angle_range, angle_range)
    matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    return cv2.warpAffine(img, matrix, (cols, rows))

# Random brightness adjustment
def brightness(img, brightness_range=50):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    brightness = np.random.randint(-brightness_range, brightness_range)
    hsv[:,:,2] = np.clip(hsv[:,:,2] + brightness, 0, 255)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
```

```python
augmented_images = []
for i in range(30): # Generate 10 augmented images
    augmented_image = horizontal_flip(img)
    augmented_image = rotation(augmented_image)
    augmented_image = brightness(augmented_image)
    augmented_images.append(augmented_image)
```

```python
for i in range(len(augmented_images)):
    plt.subplot(5,6,i+1)
    plt.imshow(augmented_images[i])
```
